```python
import os, sys
import json
import math
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns 

from os.path import join 
from sklearn import preprocessing
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder
from IPython.display import display_html

import warnings
warnings.filterwarnings("ignore")

DataPath = r'D:\GitWork\titanic\data'
OutPath  = r'D:\GitWork\titanic\output'
```

```python
f_abspath = join(DataPath, 'train.csv')
df_train = pd.read_csv(f_abspath)

display(df_train.info())
display(df_train.head())
display(df_train.describe())
```

```python
f_abspath = join(DataPath, 'test.csv')
df_test = pd.read_csv(f_abspath)

display(df_test.head())
display(df_test.describe())

display(df_test.info())
display(df_test[df_test.Fare.isnull()])

df_test.Fare.fillna(df_test.Fare.mean(), inplace=True)
display(df_test.info())
```

```python
df_data = df_train.append(df_test)

display(df_data)
display(df_data.describe())
```

```python
display(df_train.Ticket.describe())
```

```python
# Family Size
df_data['FamilySize'] = df_data.SibSp + df_data.Parch + 1
display(df_data)
```

```python
col_names = ['PassengerId', 'Name', 'Ticket', 'Fare', 'Cabin', 'FamilySize', 'Survived']

# List all Ticket counts more than 1 
dfs = []
for tk in df_data.Ticket.unique():
    tmp = df_data.loc[df_data.Ticket==tk, 'Fare']
    if tmp.count() > 1:
        dfs.append(df_data.loc[df_data.Ticket==tk, col_names])

df_tk = pd.concat(dfs)
display(df_tk.head(15))
```

```python
# df_fri  = df_tk.loc[(df_tk.FamilySize==1) & df_tk.Survived.notnull()]
# df_fami = df_tk.loc[(df_tk.FamilySize> 1) & df_tk.Survived.notnull()]
# display(df_fri.head(10), df_fami.head(10))

df_tk1 = df_tk.loc[df_tk.FamilySize == 1]
df_tk2 = df_tk.loc[df_tk.FamilySize >  1]
display(df_tk1[df_tk1.Survived.notnull()].head(10))
display(df_tk2[df_tk2.Survived.notnull()].head(10))

num_total = len(df_tk)
print('\nKeep same ticket:', len(df_tk))

num_sgl = len(df_tk1)
print('Single: {} ({}%)'.format(num_sgl, round(num_sgl/num_total, 3)))

num_mpl = len(df_tk2)
print('Family: {} ({}%)'.format(num_mpl, round(num_mpl/num_total, 3)))
```

```python
# Example

df_data['Connected_Survival'] = 0.5

for tk, df_grp in df_data.groupby('Ticket'):
    if (len(df_grp) > 1):
        for idx, row in df_grp.iterrows():
            smax = df_grp.drop(idx).Survived.max()
            smin = df_grp.drop(idx).Survived.min()
            passId = row.PassengerId
            if (smax == 1.0):
                df_data.loc[df_data.PassengerId == passId, 'Connected_Survival'] = 1
            elif (smax == 0.0):
                df_data.loc[df_data.PassengerId == passId, 'Connected_Survival'] = 0
display(df_data)                
                
# print
print('People keep same ticket: {}'.format(len(df_tk)))

num_conn_surv = df_data[df_data.Connected_Survival != 0.5]['Connected_Survival'].count()
print('People have connected infomation: {}'.format(num_conn_surv))

df_tmp = df_data.groupby('Connected_Survival')[['Survived']].mean().round(3)
display(df_tmp)
```

```python
# Set connected survival is 0.5.
# If there are someone alive in same ticket group, set connected survival is 1
# If nobody alive in same ticket group, set connected survival is 0

df_tk['Connected_Survival'] = 0.5

for tk, df_grp in df_data.groupby('Ticket'):
    if (len(df_grp) > 1):
        for idx, row in df_grp.iterrows():
            smax = df_grp.drop(idx).Survived.max()
            smin = df_grp.drop(idx).Survived.min()
            passId = row.PassengerId
            if (smax == 1.0):
                df_tk.loc[df_tk.PassengerId == passId, 'Connected_Survival'] = 1
            elif (smax == 0.0):
                df_tk.loc[df_tk.PassengerId == passId, 'Connected_Survival'] = 0
display(df_tk)
                
# print
print('People keep same ticket: {}'.format(len(df_tk)))

num_conn_surv = df_tk[df_tk.Connected_Survival != 0.5]['Connected_Survival'].count()
print('People have connected infomation: {}'.format(num_conn_surv))

ct = pd.crosstab(
    df_tk.Connected_Survival, df_tk.Survived, margins = True, margins_name='Total'
)
ct['S_Rate'] = (ct[1.0]/ ct.Total).round(3)
display(ct)
```

```python
# Training
features = ['PassengerId', 'Survived', 'Sex', 'Pclass', 'SibSp', 'Parch', 'Ticket', 'Fare']
df_dset = df_train[features]
# display(df_dset.head())

# Transform sex label to number
le = LabelEncoder()
df_dset.insert(3, 'SexCode', le.fit_transform(df_dset.Sex))
# display(df_dset.head())

def classes2dict(cls):
    return {
        le.transform([x])[0]: x for x in le.classes_
    }

dict_sex = classes2dict(le.classes_)
print('Sex dict:', dict_sex)

# Split Fare to five parts
df_dset['FareBin_5'] = pd.qcut(df_dset.Fare, 5)
df_dset['FareBinCode_5'] = le.fit_transform(df_dset.FareBin_5)
df_dset.drop(['FareBin_5'], axis=1, inplace=True)
# display(df_dset.head())

# Create connected_survival 
df_dset['FamilySize'] = df_dset.SibSp + df_dset.Parch + 1
df_dset['Connected_Survival'] = 0.5

for tk, df_grp in df_dset.groupby('Ticket'):
    if (len(df_grp) > 1):
        for idx, row in df_grp.iterrows():
            smax = df_grp.drop(idx).Survived.max()
            smin = df_grp.drop(idx).Survived.min()
            passId = row.PassengerId
            if (smax == 1.0):
                df_dset.loc[df_dset.PassengerId == passId, 'Connected_Survival'] = 1
            elif (smax == 0.0):
                df_dset.loc[df_dset.PassengerId == passId, 'Connected_Survival'] = 0

display(df_dset.head())
                
# print
print('People keep same ticket: {}'.format(len(df_dset)))

num_conn_surv = df_dset[df_dset.Connected_Survival != 0.5]['Connected_Survival'].count()
print('People have connected infomation: {}'.format(num_conn_surv))

ct = pd.crosstab(df_dset.Connected_Survival, df_dset.Survived, margins = True, margins_name='Total')
ct['S_Rate'] = (ct[1]/ ct.Total).round(3)
display(ct)

# Start training
features = ['SexCode', 'Pclass', 'FareBinCode_5', 'Connected_Survival']
inps = df_dset.filter(features, axis=1)
lbls = df_dset.Survived

model_conn = RandomForestClassifier(random_state=2, n_estimators=250, min_samples_split=20, oob_score=True)
model_conn.fit(inps, lbls)
print('oob score :%f' %(model_conn.oob_score_))
```
