```python
import os, sys
import json
import math
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import re
import seaborn as sns 

from os.path import join 
from sklearn import preprocessing
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder
from IPython.display import display_html

import warnings
warnings.filterwarnings("ignore")
```

```python
DataPath = r'D:\GitWork\titanic\data'
OutPath  = r'D:\GitWork\titanic\output'

f_abspath = join(DataPath, 'train.csv')
df_train = pd.read_csv(f_abspath)

f_abspath = join(DataPath, 'test.csv')
df_test = pd.read_csv(f_abspath)

df_data = df_train.append(df_test)
display(df_data.head())
display(df_data.describe())
display(df_data.info())
```

```python
# PPS (Predictive Power Score)
import ppscore as pps # importing ppscore

col_selected = df_data.columns.drop(['PassengerId', 'Cabin'])
df_pps = df_data[col_selected]

fig, ax = plt.subplots(figsize=(10,8)) 
matrix_df = pps.matrix(df_pps)[['x', 'y', 'ppscore']].pivot(columns='x', index='y', values='ppscore')
matrix_df = matrix_df.apply(lambda x: round(x, 2)) # Rounding matrix_df's values to 0, XX

sns.heatmap(matrix_df, vmin=0, vmax=1, cmap="Blues", linewidths=0.75, annot=True, ax=ax)
```

```python
# Show null informations
def display_side_by_side(*args):
    html_str = ''
    for df in args:
        html_str += df.to_html()
    display_html(html_str.replace('table','table style="display:inline;margin:10px"'), raw=True)
    return
```

```python
df1 = df_train.isnull().sum().to_frame().rename(columns={0:'Train null values'})
df2 = df_test.isnull().sum().to_frame().rename(columns={0:'Test null values'})
df3 = df_data.isnull().sum().to_frame().rename(columns={0:'All null values'})

display_side_by_side(df1, df2, df3)
```

```python
# Transform Sex
le_sex = LabelEncoder()
df_data['SexCode'] = le_sex.fit_transform(df_data.Sex)
print()
display(df_data.info())
display(df_data.head())

dict_sex = { le_sex.transform([x])[0]:x for x in le_sex.classes_ }
display(dict_sex)
```

```python
# Fill fare null values
pid_list = df_data[df_data.Fare.isnull()]['PassengerId'].tolist()
display(df_data.loc[df_data.PassengerId.isin(pid_list)])

mean_fare = df_data.Fare.mean().round(3)
df_data.Fare.fillna(mean_fare, inplace=True)
print('fare mean is {}'.format(mean_fare))

print('\nNumber of fare null values:', df_data.Fare.isnull().sum())
```

```python
# Fill embarked null values

df = df_data.groupby('Embarked')['Embarked'].count().reset_index(name="Count")
# df = df_data.groupby('Embarked', as_index=False)['Embarked'].agg({'Count': 'count'})
display(df)

embark_port = df.loc[df.Count.idxmax(), 'Embarked'] # Find the most embark port 'S'
print("\nThe most embarked port is '{}'".format(embark_port))

df_data.Embarked.fillna(embark_port, inplace=True) # Fill most embark port to null values
```

```python
# Fill Age null values

col_selected = ['PassengerId', 'Survived', 'Sex', 'Name', 'Age', 'SibSp', 'Parch', 'Fare', 'Pclass']
df_age = df_data[col_selected]

def getTitle(name):
    if '.' in name:
        return name.split(',')[1].split('.')[0].strip()
    else:
        return 'unknown'
    
df_age['Title'] = df_age.Name.map(lambda x: getTitle(x))

titles = df_age.Title.unique()
print('Unique titles:\n  ', sorted(titles))

df_hasage = df_age[df_age.Age.notnull()]

ct1 = pd.crosstab(
    df_hasage.Title, df_hasage.Sex, values = df_hasage.Age, aggfunc = 'median'
).fillna(0).astype(int)

tuples = [('Median', 'female'), ('Median', 'male')]
ct1.columns = pd.MultiIndex.from_tuples(tuples, names=(None, ct1.columns.name))

kwargs = { 'margins': True, 'margins_name': 'Total' }
ct2 = pd.crosstab(
    df_hasage.Title, df_hasage.Sex, values=df_hasage.Age, aggfunc=len, **kwargs
).fillna(0).astype(int)

tuples = [('Count', 'female'), ('Count', 'male'), ('Count', 'Total')]
ct2.columns = pd.MultiIndex.from_tuples(tuples, names=(None, ct2.columns.name))

display_side_by_side(ct1, ct2)
```

```python
# Split Title to 6 classes: Master, Mature (50-80), Midlife (35-50), Miss, Mr, Mrs
df_age.Title.replace(['Capt','Col'], 'Mature', inplace=True)

mask = ['Don','Dona','Dr','Jonkheer','Lady','Major','Rev','Sir']
df_age.Title.replace(mask, 'Midlife', inplace=True)

df_age.Title.replace(['Master'], 'Master', inplace=True)
df_age.Title.replace(['Miss','Mlle','Mme', 'Ms'], 'Miss', inplace=True)
df_age.Title.replace(['Mr'],  'Mr', inplace=True)
df_age.Title.replace(['Mrs','the Countess'], 'Mrs', inplace=True)

display(sorted(df_age.Title.unique()))

le_ti = LabelEncoder()
df_age['TiCode'] = le_ti.fit_transform(df_age.Title)
display(df_age.head())
```

```python
# Fill age null vlaues
print('Before fill ages:')
display(df_age.Age.describe())

age_train = df_age[df_age.Age.notnull()]
age_test  = df_age[df_age.Age.isnull()]

col_selected = ['SibSp', 'Parch', 'Fare', 'Pclass', 'TiCode']
X1 = age_train[col_selected]
y1 = age_train.Age

rfr = RandomForestRegressor(random_state=0, n_estimators=250, n_jobs=-1)
rfr.fit(X1, y1)

X2 = age_test[col_selected]
pred_ages = rfr.predict(X2).round(0)
# print(pred_ages)
df_age['TiAge'] = df_age.Age.copy()
df_age.loc[(df_age.TiAge.isnull()), 'TiAge'] = pred_ages

display(age_train.head())

print('After fill ages:')
display(df_age.TiAge.describe())
```

```python
df_data['Title'] = df_age.Title.copy()
df_data['TiCode'] = df_age.TiCode.copy()
df_data['TiAge'] = df_age.TiAge.copy()
display(df_data.isnull().sum())
```

```python
# Transform Fare
#
df_data['FareBin_5'] = pd.qcut(df_data.Fare, 5)

le_fare = LabelEncoder()
df_data['FareBinCode_5'] = le_fare.fit_transform(df_data.FareBin_5)
df_data.drop(['FareBin_5'], axis=1, inplace=True)

dict_fare = { le_fare.transform([x])[0]: x for x in le_fare.classes_}
print('Dict fare:'); display(dict_fare)
```

```python
# Add connected_survival
#
df_data['FamilySize'] = df_data.SibSp + df_data.Parch + 1
df_data['Connected_Survival'] = 0.5

for tk, df_grp in df_data.groupby('Ticket'):
    if (len(df_grp) > 1):
        for idx, row in df_grp.iterrows():
            smax = df_grp.drop(idx).Survived.max()
            smin = df_grp.drop(idx).Survived.min()
            passId = row.PassengerId
            if (smax == 1.0):
                df_data.loc[df_data.PassengerId == passId, 'Connected_Survival'] = 1
            elif (smax == 0.0):
                df_data.loc[df_data.PassengerId == passId, 'Connected_Survival'] = 0

display(df_data.head())
                
# print
print('People keep same ticket: {}'.format(len(df_data)))

num_conn = df_data[df_data.Connected_Survival != 0.5]['Connected_Survival'].count()
print('People have connected infomation: {}'.format(num_conn))

ct = pd.crosstab(df_data.Connected_Survival, df_data.Survived, margins = True, margins_name='Total')

tuples = [('Count', '0'), ('Count', '1'), ('Count', 'Total')]
ct.columns = pd.MultiIndex.from_tuples(tuples, names=(None, ct.columns.name))

ct[('Ratio', 'S1')] = (ct[('Count', '1')] / ct[('Count', 'Total')]).round(3)
display(ct)
```

```python
# Add Minor Age: If the age is less than 16 then 1 else 0
df_data['MinorAge'] = ((df_data.TiAge < 16) * 1)
display(df_data.head())

# Crosstab
kwargs = { 'margins':True, 'margins_name': 'Total' }
ct = pd.crosstab(df_data.MinorAge, df_data.Survived, **kwargs)

tuples = [('Count', '0'), ('Count', '1'), ('Count', 'Total')]
ct.columns = pd.MultiIndex.from_tuples(tuples, names=(None, ct.columns.name))

ct[('Ratio', 'S1')] = (ct[('Count', '1')] / ct[('Count', 'Total')]).round(3)
display(ct)

display(df_data.info())
```

```python
# Estimate
#
len_train = len(df_train)
ds_train = df_data[:len_train]

col_selected = ['Pclass','SexCode','FareBinCode_5','Connected_Survival','TiAge','TiCode']
inps = ds_train[col_selected]
lbls = ds_train.Survived

param_grid = {
    'random_state': [1, 2],
    'min_samples_split': [10, 20, 30, 40],
    'n_estimators': [100, 250, 500, 1000]
}

# Create a based model
rf = RandomForestClassifier()

# Instantiate the grid search model
gs = GridSearchCV(
    estimator = rf, 
    param_grid = param_grid,
    scoring='accuracy',
    cv = 3, 
    n_jobs = -1 
)

gs = gs.fit(inps, lbls)

scores = gs.cv_results_['mean_test_score']
best_score = np.max(scores)
print('Best score:', best_score)

best_params = gs.best_params_
print('\nBest params:', best_params)

# best_estimator = gs.best_estimator_
# print(); print(best_estimator)
```

```python
# Training
len_train = len(df_train)
ds_train = df_data[:len_train]

# Training
col_selected = ['Pclass','SexCode','FareBinCode_5','Connected_Survival','TiAge','TiCode']
inps = ds_train[col_selected]
lbls = ds_train.Survived

# model = RandomForestClassifier(random_state=2, n_estimators=250, min_samples_split=20, oob_score=True)

random_state = best_params['random_state']
n_estimators = best_params['n_estimators']
min_split = best_params['min_samples_split']
print(random_state, n_estimators, min_split)

model = RandomForestClassifier(
    random_state = random_state, 
    n_estimators = n_estimators, 
    min_samples_split = min_split, 
    oob_score = True
)

model.fit(inps, lbls)
print('oob score :%f' %(model.oob_score_))
```

```python
# Prediction
#
ds_test = df_data[len_train:]

col_selected = ['Pclass','SexCode','FareBinCode_5','Connected_Survival','TiAge','TiCode']
inps = ds_test[col_selected]

ds_test['Survived'] = model.predict(inps).astype(int)
display(ds_test.head())

# Save predict result to xlsx
f_abspath = join(OutPath, 'Submission_v5.csv')

features = ['PassengerId', 'Survived']
ds_test[features].to_csv(f_abspath, index=False)
```
